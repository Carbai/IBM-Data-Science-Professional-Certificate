```python
import pandas as pd
import numpy as np

df_can=pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter=2)
print('Data read into a pandas dataframe!')
```

```python
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True) ## inplace avoids copy of the dataframe
```

```python
df_can.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'}, inplace=True)
df_can.head() ## printing first lines to check that it is properly read
```

```python
df_can['Total']=df_can.sum(axis=1,numeric_only=True)
df_can.head()
```

```python
df_can.set_index('Country', inplace=True)
df_can.head()
```

```python
df_can.columns= list(map(str,df_can.columns))
years = list(map(str, range(1980,2014)))
```

```python
df_CI = df_can.loc[['India', 'China'], years] 
df_CI
```

```python
%matplotlib notebook
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_CI=df_CI.transpose()
df_CI.head()
```

```python
df_CI.plot(kind='line')
plt.title('Immigration from India and China')
plt.xlabel('Years')
plt.ylabel('Immigrants number')
plt.show()
```

```python
df_can.sort_values(by='Total',axis=0,ascending=False,inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
```

```python
df_top5.index = df_top5.index.map(int) ## change index to integer
df_top5.plot(kind='line', figsize=(14,8))

plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.show()
```

```python
df_can.sort_values(by='Total',axis=0,ascending=True,inplace=True)
df_last5 = df_can.head(5)
df_last5 = df_last5[years].transpose()
df_last5.index = df_last5.index.map(int)
df_last5.plot(alpha=0.45, kind='area', figsize=(14,8))
plt.title('Immigration Trend of the Last 5 Countries')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')
plt.show()
```

```python
ax = df_last5.plot(alpha=0.55, kind='area',stacked=False, figsize=(14,8))
ax.set_title('Immigration Trend of the Last 5 Countries')
ax.set_xlabel('Years')
ax.set_ylabel('Number of immigrants')
ax.plot()
```

Question: Use the scripting layer to display the immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013? Use an overlapping plot with 15 bins and a transparency value of 0.35.

```python
df_GAB = df_can.loc[['Greece','Albania','Bulgaria'], years]
df_GAB = df_GAB.transpose()
df_GAB.index = df_GAB.index.map(int)
count, bin_edges = np.histogram(df_GAB, 15)
df_GAB.plot(kind='hist', alpha=0.35,bins=15,figsize=(10, 6),xticks=bin_edges,
                color=['coral', 'darkslateblue', 'mediumseagreen'])

plt.title('Histogram of Immigration from Greece, Albania and Bulgaria from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants') 

plt.show()
```

Question: Using the scripting later and the df_can dataset, create a horizontal bar plot showing the total number of immigrants to Canada from the top 15 countries, for the period 1980 - 2013. Label each country with the total immigrant count.

```python
df_can.sort_values(by='Total',axis=0,ascending=True,inplace=True)
df_barh = df_can['Total'].tail(15)
df_barh
```

```python
df_barh.plot(kind='barh',figsize=(14, 10), color='steelblue')
plt.xlabel('Number of Immigrants')
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')
for index, value in enumerate(df_barh): 
    print(index, value)
    label = format(int(value), ',') # format int with commas
    plt.annotate(label, xy=(value - 147000, index - 0.10), color='white')
plt.show()
```

Question: Using a pie chart, explore the proportion (percentage) of new immigrants grouped by continents in the year 2013.

Note: You might need to play with the explore values in order to fix any overlapping slice values.



```python
explode_list = [0.0, 0, 0, 0.1, 0.1, 0.2] # ratio for each continent with which to offset each wedge.
df_continents = df_can.groupby('Continent', axis=0).sum()
```

```python
df_continents['2013'].plot(kind='pie', 
                   shadow = True,
                   autopct='%1.1f%%',
                   startangle=90,
                   labels=None,
                   pctdistance=1.12,            # the ratio between the pie center and start of text label
                   explode=explode_list)

plt.legend(df_can.index, loc='upper left')
plt.show()
```
